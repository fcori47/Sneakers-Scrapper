```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('zapatillas.xlsx')
running = df.loc[df['Categoria'] == 'Running']
running = running.loc[running['Modelo'].str.contains('Remera') == False]
running = running.loc[running['Modelo'].str.contains('Buzo') == False]
running = running.loc[running['Modelo'].str.contains('Camiseta') == False]
running = running.loc[running['Modelo'].str.contains('Campera') == False]
running = running.loc[running['Modelo'].str.contains('Chaleco') == False]
running = running.loc[running['Modelo'].str.contains('Bermuda') == False]
running = running.loc[running['Modelo'].str.contains('Gorra') == False]
running = running.loc[running['Modelo'].str.contains('Musculosa') == False]
running = running.loc[running['Modelo'].str.contains('Pantalón') == False]
running = running.loc[running['Modelo'].str.contains('Porta Celular') == False]
running = running.loc[running['Modelo'].str.contains('Riñonera') == False]
running
```

```python
moda = df.loc[df['Categoria'] == 'Moda']
moda
```

```python
training = df.loc[df['Categoria'] == 'Training']
training
```

```python
frames = [running,moda,training]
z = pd.concat(frames)
z = z.reset_index()
z.drop('index', axis=1, inplace=True)
z
```

```python
z.to_excel('zapatillas_definitivo.xlsx')
```

```python
dia = z.sample(n=400)
dia = dia.loc[dia['Cantidad'] > 0]
dia = dia.reset_index()
dia.drop('index', axis=1, inplace=True)
dia['Cantidad'] = dia['Cantidad'] * 20 / 50
dia['Cantidad'] = dia['Cantidad'].astype(int)
dia = dia.loc[dia['Cantidad'] > 0]
dia = dia.reset_index()
dia.drop('index', axis=1, inplace=True)
dia.drop('Precio 8 Cuotas', axis=1, inplace=True)
dia.drop('Precio 6 Cuotas', axis=1, inplace=True)
dia.drop('Precio 3 Cuotas', axis=1, inplace=True)
dia.rename(columns={'Cantidad': 'Vendidas'}, inplace=True)
dia.rename(columns={'Disponible en': 'Vendida en'}, inplace=True)
dia
```

```python
a = []
for i in range(1000,4700+1):
    a.append(i)
```

```python
ws = {'Codigo venta': np.random.choice(a, size=len(dia)),'Dia':pd.date_range(start='2022-01-01', end='2022-12-27'),'Modelo': dia['Modelo'], 'Marca': dia['Marca'], 'Talle': dia['Talle'], 'Color': dia['Color'],'Vendidas': dia['Vendidas'], 'Precio': dia['Precio'], 'Vendida en': dia['Vendida en'], 'Categoria': dia['Categoria']}
d = pd.DataFrame(ws)
d['Modelo'] = d['Modelo'].str.replace('adidas', 'Adidas')
d['Marca'] = d['Marca'].str.replace('adidas', 'Adidas')
d
```

```python
d.to_excel('year.xlsx')
```

```python
def afip_month_year(dataf):
    rent = 50000
    employees_low = 80000
    employees_mid = 150000
    employees_high = 200000
    services = 180000
    
    try:
        total = dataf['Vendidas*Precio'] = dataf['Vendidas'] * dataf['Precio']
        total = total.sum()
        total_iva = total - (total*0.28)
        total_paying_services = total_iva - services - rent
        total_employees_low = total_paying_services - (employees_low * 100)
        total_employees_mid = total_employees_low - (employees_mid * 80)
        total_employees_high = total_employees_mid - (employees_high * 30)
        print(f'Las ventas totales son {total}, descontando el IVA e ingresos brutos (28%) te quedaría {int(total_iva)}, pagando el alquiler/expensas te quedarian {int(total_paying_services)}')
        print(f'Después de pagarle a todos los empleados te quedarían {int(total_employees_high)}')
        
    except Exception as e:
        print(f'Error: {e}')
```

```python
afip_month_year(d)
```
