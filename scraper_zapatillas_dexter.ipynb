```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
def sneakers(soup):
    url_products = 'https://www.dexter.com.ar/'
    
    try: 
        finding_links = soup.find('div', attrs= {'class': 'row product-grid'}).find_all('a')
        if finding_links:
            finding_links = finding_links
        else:
            return None
    
        
        all_products = [i.get('href') for i in finding_links]
        if all_products:
            products_links = []
            for z in all_products:
                if z == '/on/demandware.store/Sites-Dexter-Site/default/Wishlist-AddProduct' or len(z) > 80:
                    continue
                else:
                    products_links.append(url_products + z)
        else:
            return None
    
    except Exception as e:
        print(f'Error {e}')
        
    products_links = list(set(products_links))
    return products_links
```

```python
sneakers(BeautifulSoup(requests.get('https://www.dexter.com.ar/categorias/training?start=0').text, 'lxml'))
```

```python
all_url = []
for i in range(1296+1):
    if i % 36 == 0:
        a = f'https://www.dexter.com.ar/categorias/training?start={i}'
        all_url.append(a)
print(all_url)
```

```python
def extract_info_sneakers(soup):
    sneakers = {}
    try:
        name = soup.find('h1', attrs= {'class': 'product-name'}).text
        if name:
            sneakers['name'] = name
        else:
            sneakers['name'] = None
    
        price = soup.find('span', attrs = {'class': 'value'}).text
        price = price.replace(' ', '')
        price = price.replace('\n', '')
                
        if price:
            sneakers['price'] = price
        else:
            sneakers['price'] = None
       
        try:
            brand = soup.select_one('span:-soup-contains("Marca:")').find_next_sibling(text=True)
            if brand:
                sneakers['brand'] = brand.strip()
            else:
                sneakers['brand'] = None
        except Exception as e:
            return None
        
        return sneakers
    
    except Exception as e:
        print(f'Error: {e}')
```

```python
extract_info_sneakers(BeautifulSoup(requests.get('https://www.dexter.com.ar/zapatillas-topper-squat/TO_27173.html').text, 'lxml'))
```

```python
def scrape(url):
    try:
        sneaker = requests.get(url)
    except Exception as e:
        print(f'Error: {e} whit {url}')
        return None
    
    if sneaker.status_code != 200:
        print(f'Error whit this sneaker {url}')
        print(f'Error: {url.status_code}')
        return None
    
    soup_sneaker = BeautifulSoup(sneaker.text, 'lxml')
    sneakers = extract_info_sneakers(soup_sneaker)
    return sneakers
```

```python
all_pages_links = []

for i in all_url:
    try:
        iuju = sneakers(BeautifulSoup(requests.get(i).text, 'lxml'))
        all_pages_links.append(iuju)
        all_pages_links
    except Exception as e:
        None
```

```python
clean_links = []
for i in all_pages_links:
    for x in i:
        meper = x
        clean_links.append(meper)
```

```python
clean_links
```

```python
sneakers_data = []
for i, sneaker in enumerate(clean_links):
    print(f'Sneaker {i+1}/{len(clean_links)}')
    sneakers_data.append(scrape(sneaker))
```

```python
filtered_sneakers = list(filter(None,sneakers_data))
df = pd.DataFrame(filtered_sneakers)
df
```

```python
zapatillas_moda = {'Modelo': df.name,'Marca': df.brand, 'Promociones': '6/8 Sin interes (Naranja X), 3 Sin interes (Nike)', 'Precio': df.price}
moda = pd.DataFrame(zapatillas_moda)
moda['Precio'] = moda['Precio'].str.replace('$', '', regex=True)
moda['Precio'] = moda['Precio'].str.replace('.', '', regex=True)
moda['Precio'] = moda['Precio'].astype(int)
moda['Precio 8 Cuotas'] = moda['Precio'] / 8
moda['Precio 6 Cuotas'] = moda['Precio'] / 6
moda['Precio 3 Cuotas'] = moda['Precio'] / 3
moda['Precio 8 Cuotas'] = moda['Precio 8 Cuotas'].round(2)
moda['Precio 6 Cuotas'] = moda['Precio 6 Cuotas'].round(2)
moda['Precio 3 Cuotas'] = moda['Precio 3 Cuotas'].round(2)
colores = ['Negro','Rojo', 'Verde', 'Rosa', 'Azul', 'Violeta', 'Amarillo']
moda['Color'] = np.random.choice(colores, size=len(moda))

moda
```

```python
moda_test = moda.copy()
moda_test = moda_test.drop_duplicates()
moda_test['Cantidad'] = np.array(moda.value_counts())
locales = ['Nike Store Alto Palermo','Nike Store Abasto','Nike Store Alcorta', 'Nike Store DOT', 'Nike Store Solar', 'Nike Store Avenida Santa Fe', 'Nike Store Tortugas', 'Nike Store Pilar', 'Nike Store Unicenter', 'Nike Store Avellaneda', 'Nike Store Plaza Oeste']
moda_test['Disponible en'] = np.random.choice(locales, size=len(moda_test))
moda_test = moda_test.reset_index()
moda_test.drop('index', axis=1, inplace= True)
moda_test
```

```python
talles = [35,36,37,38,39,40,41,42]

ws = {'Modelo': moda_test['Modelo'], 'Marca': moda_test['Marca'], 'Talle': np.random.choice(talles, size=len(moda_test)), 'Color': moda_test['Color'], 'Cantidad': np.random.randint(30, size=len(moda_test))  ,'Precio': moda_test['Precio'], 'Precio 8 Cuotas': moda_test['Precio 8 Cuotas'], 'Precio 6 Cuotas': moda_test['Precio 6 Cuotas'], 'Precio 3 Cuotas': moda_test['Precio 3 Cuotas'],'Disponible en': moda_test['Disponible en']   }
m = pd.DataFrame(ws)
m
```
